## Analysis of Algorithms

#### Merge Sort 

For ascending, descending, and random arrays, the size of the array affects the time by approximately $nlog(n)$, where n is the size of the array.  The intial ordering of the array did not affect the time at all.  The ascending, descending, and random arrays ran in approximately the same time at each array size.

```python
print('Merge Sort Data\n')
print(mergeData, '\n')
```

#### Bubble Sort

For random and descending arrays, the size of the array affects the time by approximately $n^2$, where n is the size of the array.  And for ascending arrays, the size of the array affects the time by approximately $n$.  The difference is due to bubble sort checking at the end of iteration to see if the array is sorted.  If the array is intially sorted, then it only needs to iterate once through the array, causing it to finish in $n$ time.  Another thing to note is that for bubble sort, random arrays are slower than descending arrays, which is odd due to random arrays have the possibility to be sorted and finish before $n^2$ iterations, while descending do not. Descending arrays should take more time to sort than random arrays. However this is not true due to branch prediction.  Descending array's branch prediction will always be the same, so the computer can optimize the code, while random array's branch prediction will not always be right, so it will be slower.

```python
print('Bubble Sort Data\n')
print(bubbleData, '\n')
```

#### Selection Sort

For ascending, descending, and random arrays, the size of the array affects the time by approximately $n^2$, where n is the size of the array.  The ascending, descending, and random arrays ran in approximately the same time at each array size.  The times for all of arrays are the same because there is no check to see when the array becomes sorted and no branch prediction differences, like in bubble sort.

```python
print('Selection Sort Data\n')
print(selectionData)
```

#### Comparing the Sorting Algorithms

For random or descending arrays, merge sort was the fastest due to size affecting it's time by $nlog(n)$.  However, if the array was already sorted (ascending), then bubble sort was the fastest by far due to its ability to finish in $n$ time.  Additionally, selection sort was faster than bubble sort  at sorting random arrays due to it only iterating the unsorted part of the array instead of the entire array, like in bubble sort.  The ability for bubble sort to finish early did not make a big enough difference to matter in random arrays. However, for descending arrays, bubble sort was faster than selection due to branch predicition optimization.
